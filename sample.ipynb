```python
import matplotlib.pyplot as plt
```

# Ethos

```python
baseline_acc = [0.8, 0.84, 0.82, 0.88]
new_acc = [0.8, 0.89, 0.95, 0.92]
baseline_val = [0.8, 0.84, 0.82, 0.88]
new_val = [0.8, 0.86, 0.88, 0.9]
baseline_time = [2.5, 174.68, 920.221, 1273.025]
new_time = [1.6, 71.11, 352.9, 515.64]
```

```python
epochs = list(range(1, 5))

# Create a figure with 3 subplots in 1 row
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # width 18, height 5

# Plot 1: Accuracy
axes[0].plot(epochs, baseline_acc, label="Baseline Accuracy")
axes[0].plot(epochs, new_acc, label="New Accuracy")
axes[0].set_xlabel("Round")
axes[0].set_ylabel("Accuracy")
axes[0].set_title("Training Accuracy")
axes[0].legend()

# Plot 2: Validation Accuracy
axes[1].plot(epochs, baseline_val, label="Baseline Validation Accuracy")
axes[1].plot(epochs, new_val, label="New Validation Accuracy")
axes[1].set_xlabel("Round")
axes[1].set_ylabel("Validation Accuracy")
axes[1].set_title("Validation Accuracy")
axes[1].legend()

# Plot 3: Time per Epoch
axes[2].plot(epochs, baseline_time, label="Baseline Time")
axes[2].plot(epochs, new_time, label="New Time")
axes[2].set_xlabel("Round")
axes[2].set_ylabel("Time (s)")
axes[2].set_title("Training Time per Epoch")
axes[2].legend()

plt.tight_layout()  # Adjust spacing
plt.show()
```
